### Importing and reading required data

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Cars Manufacturing.csv")
data.head()
```

```python
del data["Unnamed: 0"]
del data["vin"]
del data["lot"]
del data["condition"]
```

### Q1. Top 5 Car brands in terms of number of cars

```python
data_cb = []

for brand in data['brand'].unique():
    data_cb.append([brand , len(data[data['brand']  == brand])])

Car_brand = pd.DataFrame(data_cb,columns = ['brand','freq']).sort_values(by='freq', ascending=False).head()
Car_brand
```

```python
# Plotting bar graph
Car_brand.plot(kind="bar", x = 'brand', y = 'freq', figsize = (9, 5), fontsize = 10)
plt.title("Top 5 Car brands in terms of number of cars", fontsize = 16)
plt.xlabel("Brand of Cars", fontsize = 13)
plt.ylabel("Number of Cars", fontsize = 13)
plt.show()
```

### Q.2.Car brand based on Frequency of Cars

```python
Car_brand_f = pd.DataFrame(data_cb,columns = ['brand','freq']).sort_values(by='freq', ascending=False)
Car_brand_f
```

```python
# Plotting bar graph
Car_brand_f.plot(kind="bar", x = 'brand', y = 'freq', figsize = (15, 5), fontsize = 10)
plt.title("Car brand based on Frequency of Cars", fontsize = 16)
plt.xlabel("Brand of Cars", fontsize = 13)
plt.ylabel("Number of Cars", fontsize = 13)
plt.show()
```

### 3. Car Industry Trend

```python
data_it = []

for year in data['year'].unique():
    data_it.append([year , len(data[data['year']  == year])])

Car_industry_trend = pd.DataFrame(data_it,columns = ['year','trend']).sort_values(by='year', ascending = True)
Car_industry_trend
```

```python
# Plotting line graph
Car_industry_trend.plot(kind="line", x = 'year', y = 'trend', figsize = (15, 5), fontsize = 10)
plt.title("Car Industry Trend", fontsize = 16)
plt.xlabel("Years", fontsize = 13)
plt.ylabel("Industry Trend", fontsize = 13)
plt.show()
```

### Q.4. Most Populat Car Colours

```python
data_cpc = []

for color in data['color'].unique():
    data_cpc.append([color , len(data[data['color']  == color])])

Populate_car_color = pd.DataFrame(data_cpc,columns = ['color','car_no']).sort_values(by='car_no', ascending=False)
Populate_car_color
```

```python
# Plotting bar graph
Populate_car_color
```

### Q5. Which car brand is having Maximum Average Price

```python
Avg_price = data[["brand", "price"]].groupby("brand").mean().sort_values(by = "price", ascending = False)
Avg_price
```

```python
Avg_price.plot(kind="bar",figsize = (15, 5), fontsize = 10)
plt.title("Brand having Max. Average Price", fontsize = 16)
plt.xlabel("Brand", fontsize = 13)
plt.ylabel("Avg. Price", fontsize = 13)
plt.show()
```

### Q6. Which State is having most expensive cars

```python
State = data[["state", "price"]].groupby("state").mean().sort_values(by = "price", ascending = False)
State
```

```python
# Plotting bar graph
State.plot(kind="bar",figsize = (15, 5), fontsize = 10)
plt.title("State having Most Expensive Car", fontsize = 16)
plt.xlabel("State", fontsize = 13)
plt.ylabel("Car Price", fontsize = 13)
plt.show()
```
